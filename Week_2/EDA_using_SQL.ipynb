### Task 1




##### Display the names of the unique launch sites  in the space mission


```python
print(df["Launch_Site"].unique())
```


### Task 2


#####  Display 5 records where launch sites begin with the string 'CCA'


```python
count = 0
inds = []
for i in range(len(df["Launch_Site"])):
  if df["Launch_Site"][i][:3] == "CCA":
    inds.append(i)
    count += 1
  if count == 5: break
```

```python
df.iloc[[k for k in inds]]
```

### Task 3




##### Display the total payload mass carried by boosters launched by NASA (CRS)


```python
sum([df["PAYLOAD_MASS__KG_"][k] for k in range(len(df)) if df["Customer"][k] == "NASA (CRS)"])
```

### Task 4




##### Display average payload mass carried by booster version F9 v1.1


```python
sum([df["PAYLOAD_MASS__KG_"][k] for k in range(len(df)) if df["Booster_Version"][k] == "F9 v1.1"])/len(df)
```

### Task 5

##### List the date when the first succesful landing outcome in ground pad was acheived.


_Hint:Use min function_


```python
min([df["Date"][k] for k in range(len(df)) if df["Landing_Outcome"][k] == "Success (ground pad)"])
```

### Task 6

##### List the names of the boosters which have success in drone ship and have payload mass greater than 4000 but less than 6000


```python
min(set([df["Booster_Version"][k] for k in range(len(df)) if df["Landing_Outcome"][k] == "Success (drone ship)" and 4000 < df["PAYLOAD_MASS__KG_"][k] < 6000]))
```

### Task 7




##### List the total number of successful and failure mission outcomes


```python
len(df)
```

```python
k = len([df["Mission_Outcome"][j] for j in range(len(df)) if type(df["Mission_Outcome"][j]) == str and df["Mission_Outcome"][j][:7] == "Success"])
print("Number of sucessful missions: ", k)
print("Number of failure missions: ", len(df) - k)
```

### Task 8



##### List the   names of the booster_versions which have carried the maximum payload mass. Use a subquery


```python
max_payload = max(df["PAYLOAD_MASS__KG_"])
```

```python
set([df["Booster_Version"][k] for k in range(len(df)) if df["PAYLOAD_MASS__KG_"][k] == max_payload])
```
